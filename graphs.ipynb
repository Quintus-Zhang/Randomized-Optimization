## Continuous Peaks

### 1. fitness graph

```python
fit_rhc = pd.read_csv('cp_fitness_rhc.csv', header=None)
fit_sa = pd.read_csv('cp_fitness_sa.csv', header=None)
fit_ga = pd.read_csv('cp_fitness_ga.csv', header=None)
fit_mimic = pd.read_csv('cp_fitness_mimic.csv', header=None)

fit_rhc_mean = fit_rhc.mean()
fit_rhc_std = fit_rhc.std()
fit_sa_mean = fit_sa.mean()
fit_sa_std = fit_sa.std()
fit_ga_mean = fit_ga.mean()
fit_ga_std = fit_ga.std()
fit_mimic_mean = fit_mimic.mean()
fit_mimic_std = fit_mimic.std()


plt.figure()
plt.grid(linewidth=1)
plt.xlabel("Iterations in hundreds")
plt.ylabel("Fitness")
plt.title('Continuous Peaks: fitness over iterations')

# RHC
plt.fill_between(fit_rhc.columns, fit_rhc_mean - fit_rhc_std,
                 fit_rhc_mean + fit_rhc_std, alpha=0.1,
                 color="r")
plt.semilogx(fit_rhc.columns, fit_rhc_mean, '-', color="r",
         label="RHC")
# SA
plt.fill_between(fit_sa.columns, fit_sa_mean - fit_sa_std,
                 fit_sa_mean + fit_sa_std, alpha=0.1,
                 color="b")
plt.semilogx(fit_sa.columns, fit_sa_mean, '-', color="b",
         label="SA")
# GA
plt.fill_between(fit_ga.columns, fit_ga_mean - fit_ga_std,
                 fit_ga_mean + fit_ga_std, alpha=0.1,
                 color="g")
plt.semilogx(fit_ga.columns, fit_ga_mean, '-', color="g",
         label="GA")
# MIMIC
plt.fill_between(fit_mimic.columns, fit_mimic_mean - fit_mimic_std,
                 fit_mimic_mean + fit_mimic_std, alpha=0.1,
                 color="y")
plt.semilogx(fit_mimic.columns, fit_mimic_mean, '-', color="y",
         label="MIMIC")

plt.legend(loc="best")
plt.show()
```

```python
plt.figure()
plt.xlabel("Iterations in hundreds")
plt.ylabel("Standard Deviation")
plt.title('Continuous Peaks: standard deviation over iterations')

plt.plot(fit_rhc_std, label='RHC')
plt.plot(fit_sa_std, label='SA')
plt.plot(fit_ga_std, label='GA')
plt.plot(fit_mimic_std, label='MIMIC')
plt.legend()
plt.show()
```

### 2. clock time graph

```python
t_rhc = pd.read_csv('cp_time_rhc.csv', header=None)
t_sa = pd.read_csv('cp_time_sa.csv', header=None)
t_ga = pd.read_csv('cp_time_ga.csv', header=None)
t_mimic = pd.read_csv('cp_time_mimic.csv', header=None)

t_rhc_mean = t_rhc.mean()
t_rhc_std = t_rhc.std()
t_sa_mean = t_sa.mean()
t_sa_std = t_sa.std()
t_ga_mean = t_ga.mean()
t_ga_std = t_ga.std()
t_mimic_mean = t_mimic.mean()
t_mimic_std = t_mimic.std()


plt.figure()
plt.grid(linewidth=1)
plt.xlabel("Iterations in hundreds")
plt.ylabel("Clock Time in seconds")
plt.title('Continuous Peaks: clock time in log scale')

# RHC
plt.loglog(t_rhc.columns, t_rhc_mean, '-', color="r",
         label="RHC")
# SA
plt.loglog(t_sa.columns, t_sa_mean, '-', color="b",
         label="SA")
# GA
plt.loglog(t_ga.columns, t_ga_mean, '-', color="g",
         label="GA")
# MIMIC
plt.loglog(t_mimic.columns, t_mimic_mean, '-', color="y",
         label="MIMIC")

plt.legend(loc="best")
plt.show()
```
